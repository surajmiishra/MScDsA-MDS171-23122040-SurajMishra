```python
matrix_dict = {
    'matrix 1': [],
    'matrix 2': [],
    'result matrix': []
}

def collect_matrix(matrix_name):
    print(f"Enter {matrix_name}:")
    rows, cols = map(int, input("Enter the number of rows and columns (e.g., 2 3): ").split())
    matrix = [[float(input(f"Enter the value at row {i + 1}, column {j + 1}: ")) for j in range(cols)] for i in range(rows)]
    matrix_dict[matrix_name] = matrix
    print(f"{matrix_name} collected successfully.")

def multiply_matrices():
    matrix1, matrix2 = matrix_dict['matrix 1'], matrix_dict['matrix 2']
    if len(matrix1[0]) != len(matrix2):
        print("Matrix multiplication is not possible. The number of columns in matrix 1 must be equal to the number of rows in matrix 2.")
        return
    result = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(len(matrix2))) for j in range(len(matrix2[0]))] for i in range(len(matrix1))]
    matrix_dict['result matrix'] = result
    print("Matrix multiplication completed successfully.")

def export_result_to_file(filename):
    result_matrix = matrix_dict['result matrix']
    with open(filename, 'w') as file:
        file.writelines([' '.join(map(str, row)) + '\n' for row in result_matrix])
    print(f"Result matrix exported to {filename} successfully.")

while True:
    print("\nMenu:")
    print("1. Enter Matrix 1")
    print("2. Enter Matrix 2")
    print("3. Calculate")
    print("4. Export Result to File")
    print("5. Quit")
    choice = input("Enter your choice (1/2/3/4/5): ")
    
    if choice == '1':
        collect_matrix('matrix 1')
    elif choice == '2':
        collect_matrix('matrix 2')
    elif choice == '3':
        multiply_matrices()
    elif choice == '4':
        filename = input("Enter the filename to export the result matrix: ")
        export_result_to_file(filename)
    elif choice == '5':
        break
    else:
        print("Invalid choice. Please select a valid option.")
```
