### Pre-Processing

```python
df.dropna(subset=['Geographic Area', 'Residence'], inplace=True)
```

```python
df['Geographic Area'].fillna('Unknown', inplace=True)
```

```python
df=df.drop(["Timestamp","Email Address"],axis=1)
```

```python
average1 = int(2000//2)
average2 = int((2000+4000)//2)
average3 = int((4000+6000)//2)
average4 = int((6000+8000)//2)
average5 = int((8000)//2)
df["Expenditure for Online Shopping" ]=df["Expenditure for Online Shopping"].replace('0-₹2000', average1)
df["Expenditure for Online Shopping" ]=df["Expenditure for Online Shopping"].replace('₹2000-₹4000', average2)
df["Expenditure for Online Shopping" ]=df["Expenditure for Online Shopping"].replace('₹4000-₹6000', average3)
df["Expenditure for Online Shopping" ]=df["Expenditure for Online Shopping"].replace('₹6000-₹8000', average4)
df["Expenditure for Online Shopping" ]=df["Expenditure for Online Shopping"].replace('More than 8000', average5)
```

```python
averagee1 = int(5000//2)
averagee2 = int((5000+10000)//2)
averagee3 = int((4000+6000)//2)
averagee4 = int((6000+8000)//2)
averagee5 = int((20000)//2)
df["Monthly Allowance" ]=df["Monthly Allowance"].replace('0-5000', averagee1)
df["Monthly Allowance" ]=df["Monthly Allowance"].replace('5000-10000', averagee2)
df["Monthly Allowance" ]=df["Monthly Allowance"].replace('10000-15000', averagee3)
df["Monthly Allowance" ]=df["Monthly Allowance"].replace('15000-20000', averagee4)
df["Monthly Allowance" ]=df["Monthly Allowance"].replace('More than 20000', averagee5)
```

```python

df["Frequency of Shopping" ]=df["Frequency of Shopping"].replace("Daily",1)
df["Frequency of Shopping" ]=df["Frequency of Shopping"].replace("Weekly",2)
df["Frequency of Shopping" ]=df["Frequency of Shopping"].replace("Monthly",3)
df["Frequency of Shopping" ]=df["Frequency of Shopping"].replace("Rarely",4)
df["Frequency of Shopping" ]=df["Frequency of Shopping"].replace("Never",5)
```

```python
df["Gender"] = df["Gender"].replace("Male",0)
df["Gender"] = df["Gender"].replace("Female",1)
```

## Analysis

### The Satisfaction of Shopping Experience with respect to Gender Analysis 

```python
male_data = df[df['Gender'] == 'Male']
female_data = df[df['Gender'] == 'Female']
male_satisfaction = male_data['Satisfaction Level for Shopping Online']
female_satisfaction = female_data['Satisfaction Level for Shopping Online']

plt.figure(figsize=(10, 4))
plt.hist(male_satisfaction, bins=10, alpha=0.6, label='Male')
plt.hist(female_satisfaction, bins=10, alpha=0.6, label='Female')
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.title('Satisfaction Level Histogram by Gender')
plt.legend()
plt.show()

```

### Insight
- Here we are analyzing the the male and female satisfication on an overall basis, Hence here the Female Satisfaction is more when we try to take the average with respect to the higher satisfaction level. So now lets try to explore one what all basis this satisfaction could be analyzed.

Explore whether age, gender affect the selection of e-commerce platforms. Are there certain platforms that are more popular among specific demographic groups?

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))
sns.countplot(x='Age', hue='Indian E-Commerce Platform(Used)', data=df)
plt.xlabel('Age')
plt.ylabel('Count') #The count here is nothing but the occurances of various age groups from the dataframe. Whereas the hue looks into the Platforms. 
plt.title('Platform Usage by Age Group')
plt.xticks(rotation=45)
plt.legend(title='Platform')
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.countplot(x='Improvements', hue='Indian E-Commerce Platform(Used)', data=df)
plt.xlabel('Improvements')
plt.ylabel('Count')
plt.title('Comparison of Improvements by E-Commerce Platform')
plt.xticks(rotation=45)
plt.legend(title='Platform')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group the data by age and platform choice and calculate the mean expenditure
age_platform_mean_expenditure = df.groupby(['Age', 'Indian E-Commerce Platform(Used)'])['Expenditure for Online Shopping'].mean().unstack()

# Create a bar chart
age_platform_mean_expenditure.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age Group')
plt.ylabel('Mean Expenditure')
plt.title('Mean Expenditure on E-commerce Platforms by Age Group')
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### **Hostel Residence and Satisfaction with Services**:
Investigate whether the hostel where participants reside has any correlation with their
satisfaction levels regarding e-commerce services. Are there hostels that consistently receive
higher or lower satisfaction ratings?

```python
import matplotlib.pyplot as plt

residence_satisfaction_counts = df.groupby(['Residence', 'Satisfaction Level for Shopping Online'])["Age"].count().unstack()

ax = residence_satisfaction_counts.plot(kind='bar', figsize=(12, 6))
plt.xlabel('Residence')
plt.ylabel('Count')
plt.title('Distribution of Satisfaction Levels for Shopping Online by Residence')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Satisfaction Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Insights:

- The Hostel residents prefer less shopping compared with the private. When it comes to monthly and Rarely the count is high for Private residents.
